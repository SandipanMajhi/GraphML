## Main Methodology

We employ multistage prompting by breaking down a bigger task to smaller ones. We find ask the LLM to get the adjacency list representation of the graph, then ask it to give a detailed overview of the algorithm it is going to use. Then finally do a chain-of-thought prompting to follow the algorithm step by step.

Please - Input graphs in proper format with no extra spaces and extra new lines.

# Find existance of cycle in a graph

*Task Definition:* Given input undirected graph, the task is to check whether the graph contains a cycle or not.

Algorithm Hint - https://www.geeksforgeeks.org/detect-cycle-in-a-graph/

Important Note: Please provide the input in the following format only. We kind of parse the data and then do multistage prompting.
```
    Q_cycle = """
    Graph: (0,2),(2,4),(4,6),(6,8),(8,9),(9,7),(7,5),(5,3),(3,1)
    Q: Is there a cycle in the graph?
    """
```

````python
Q_cycle = """
Graph: (0,2),(2,4),(4,6),(6,8),(8,9),(9,7),(7,5),(5,3),(3,1)
Q: Is there a cycle in the graph?
"""
````

````python
graph_extraction_prompt = """Given the undirected graph express the graph in adjacency list format.

Graph: (0,1) (0,2) (3,4) (4,5)
Answer: 0:[1,2], 1:[0], 2:[0], 3:[4], 4:[5], 5:[4]
"""

cycle_algo_prompt = """
Given the above question mention step by step in detail needed to solve the question. Please do not provide examples. Only steps in detail.
Answer:"""

cycle_solution_prompt = """
"""

question = Q_cycle.split("\n")
question = [q for q in question if len(q) > 1]

graph_extraction_prompt = f"""{graph_extraction_prompt}
{question[0]}
Answer:"""

cycle_algo_prompt = question[1] + cycle_algo_prompt
````

````python
client = Groq()
graph_completion = client.chat.completions.create(
    model="llama-3.2-90b-text-preview",
    messages=[
        {
            "role": "user",
            "content": graph_extraction_prompt
        }
    ],
    temperature=0.5,
    max_tokens=8192,
    top_p=1,
    stream=False,
    stop=None,
)

algo_completion = client.chat.completions.create(
    model="llama-3.2-90b-text-preview",
    messages=[
        {
            "role": "user",
            "content": cycle_algo_prompt
        }
    ],
    temperature=0.5,
    max_tokens=8192,
    top_p=1,
    stream=False,
    stop=None,
)


cycle_solution_prompt = f"""Graph Adjacency List: {graph_completion.choices[0].message.content}
Algorithm: {algo_completion.choices[0].message.content}
Given the above Graph Adjacency List use the above algorithm for existance of cycles give step by step solution.
"""

sol_completion = client.chat.completions.create(
    model="llama-3.2-90b-text-preview",
    messages=[
        {
            "role": "user",
            "content": cycle_solution_prompt
        }
    ],
    temperature=0.5,
    max_tokens=8192,
    top_p=1,
    stream=False,
    stop=None,
)
````

````python
#### Graph Adjacency List Extraction LLM Answer #####
display(Markdown(graph_completion.choices[0].message.content))
````

````python
##### Cycle Algorithm LLM answer ####
display(Markdown(algo_completion.choices[0].message.content))
````

````python
#### Final solution to the given graph if cycle present or not ####
display(Markdown(sol_completion.choices[0].message.content))
````
